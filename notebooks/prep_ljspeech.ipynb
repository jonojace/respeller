```python
# notebook to prep a freshly downloaded ljspeech for processing using this fastpitch repo

# wget https://data.keithito.com/data/speech/LJSpeech-1.1.tar.bz2
# tar -xvf LJSpeech-1.1.tar.bz2 -C ./wavs

import os

wav_dir = '/home/s1785140/data/ljspeech_fastpitch/wavs'
metadata = '/home/s1785140/data/ljspeech_fastpitch/metadata.csv'
outdir = '/home/s1785140/data/ljspeech_fastpitch'
```

```python
# for creating train_meta and dev_meta files
train_chapters = list(range(1,25))
dev_chapters = [25]

with open(metadata, 'r') as f:
    lines =f.readlines()
    
lines_to_write = []
train_meta_lines = []
dev_meta_lines = []
for l in lines:
    uttid, raw_text, norm_text = l.split('|')
    norm_text = norm_text.lower()
    
    path_to_wav = os.path.join(wav_dir, f"{uttid}.wav")
    new_line = f"{path_to_wav}|{norm_text}"
    lines_to_write.append(new_line)
    
    meta_line = f"mels/{uttid}.pt|durations/{uttid}.pt|pitches/{uttid}.pt|{norm_text}"
    chapter = int(uttid.split('-')[0].lstrip('LJ0'))
    if chapter in train_chapters:
        train_meta_lines.append(meta_line)
    elif chapter in dev_chapters:
        dev_meta_lines.append(meta_line)
    
with open(os.path.join(outdir, 'wav_text_filelist.txt'), 'w') as f:
    f.writelines(lines_to_write)
    
with open(os.path.join(outdir, 'train_meta.txt'), 'w') as f:
    f.writelines(train_meta_lines)
    
with open(os.path.join(outdir, 'val_meta.txt'), 'w') as f:
    f.writelines(dev_meta_lines)
```
